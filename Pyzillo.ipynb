```python
import time
import pandas as pd
import simplejson
import urllib
import requests
from bs4 import BeautifulSoup
```

```python
def fetch_search_results(query=None, minAsk=None, maxAsk=None, bedrooms=None, number=0):
    search_params = {
        key: val for key, val in locals().items() if val is not None
    }
    if not search_params:
        raise ValueError("No valid keywords")

    base = 'https://sfbay.craigslist.org/search/sby/apa?s=%d' %number
    resp = requests.get(base, params=search_params, timeout=3)
    resp.raise_for_status()  # <- no-op if status==200
    return resp.content, resp.encoding
```

```python
def extract_listings(parsed):
    listings = parsed.find_all('p', class_='row')
    return listings
```

```python
def parse_source(html, encoding='utf-8'):
    parsed = BeautifulSoup(html, from_encoding=encoding)
    return parsed
```

```python
def pages(num):
    html, encoding = fetch_search_results(minAsk=2000, maxAsk=2500, bedrooms=2, number=num)
    doc = parse_source(html, encoding)
    listings = extract_listings(doc)           
    return listings
```

```python
def distance(mode, original_cord, destination_cord):
    url = ("http://maps.googleapis.com/maps/api/distancematrix/json?origins={0}&destinations={1}&mode={2}&language=en-EN&sensor=false".format
           (str(original_cord),str(destination_cord), str(mode)))
    result= simplejson.load(urllib.urlopen(url))
    driving_time_text = result['rows'][0]['elements'][0]['duration']['text']
    driving_time_value = result['rows'][0]['elements'][0]['duration']['value']
    return driving_time_text, driving_time_value
```

```python
samsung = (37.3876191,-121.9324151)
dixon_landing = (37.4544319,-121.9204801)
cgate = 37.480636,-121.9406515
distance('bicycling', samsung, dixon_landing)
```

```python
def attributes(s):
    t1 = map(lambda x: x.strip(), s.split('/'))
    bedrooms = float(t1[0].split('BR')[0])
    bathrooms = float(t1[1].split('Ba')[0])
    t2 = t1[1].split('available')
    t3 = t2[0].split('Ba')
    available_date = t2[1].strip()    
    area = t3[1].strip()
    return bedrooms, bathrooms, available_date, area
```

```python
list_bedrooms = []
list_bathrooms = [] 
list_available_date = [] 
list_area = [] 
list_other_attributes = []
list_samsung_driving_distance = []
list_samsung_bicycling_distance = []
list_cgate_driving_distance = []
list_cgate_bicycling_distance = []
list_price = []
list_url = []

for num in xrange(0, 2500, 100):
    listings = pages(num)
    for element in listings:
        href = element.a['href']
        url = 'http://sfbay.craigslist.org' + href
        resp = requests.get(url,  timeout=10)
        time.sleep(2)
        html, encoding = resp.content, resp.encoding
        parsed = BeautifulSoup(html, from_encoding=encoding)
        t = [a.get_text() for a in parsed.find_all('p', class_="attrgroup")]

        bedrooms = 0
        bathrooms = 0 
        available_date = ''
        area = ''
        other_attributes = ''
        distance_samsung_bicycling = 0
        distance_samsung_driving = 0
        distance_cgate_bicycling = 0
        distance_cgate_driving = 0
        price = 0
        try:
            bedrooms, bathrooms, available_date, area = attributes(t[0])
            other_attributes = ', '.join(t[1:])
        except:
            pass
        try:
            longitude = parsed.find_all('div', class_='viewposting')[0]['data-longitude']
            latitude = parsed.find_all('div', class_='viewposting')[0]['data-latitude']
            location = (float(latitude), float(longitude))
            distance_samsung_bicycling = distance('bicycling', location, samsung)
            distance_samsung_driving = distance('driving', location, samsung)
            distance_cgate_bicycling = distance('bicycling', location, cgate)
            distance_cgate_driving = distance('driving', location, cgate)
            price = parsed.find_all('span', class_='price')[0].get_text()
        except:
            pass


        list_bedrooms.append(bedrooms)
        list_bathrooms.append(bathrooms)
        list_available_date.append(available_date)
        list_area.append(area)
        list_other_attributes.append(other_attributes)
        list_samsung_driving_distance.append(distance_samsung_driving)
        list_samsung_bicycling_distance.append(distance_samsung_bicycling)
        list_cgate_driving_distance.append(distance_cgate_driving)
        list_cgate_bicycling_distance.append(distance_cgate_bicycling)
        list_price.append(price)
        list_url.append(url)
        time.sleep(10)
```

```python
df = pd.DataFrame({'bedrooms': list_bedrooms, 'bathrooms' : list_bathrooms,
                   'price': list_price,
                   'area': list_area, 'other_attributes': list_other_attributes,
                   'distance_samsung_bicycling': list_samsung_bicycling_distance,
                   'distance_cgate_bicycling': list_cgate_bicycling_distance,
                   'distance_samsung_driving': list_samsung_driving_distance,
                   'distance_cgate_driving': list_cgate_driving_distance,
                   'url':list_url})
```
